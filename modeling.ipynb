# Clustering with 3 features

```python
dummy_df = pd.get_dummies(df[['species']], dummy_na=False)
df = pd.concat([df, dummy_df], axis=1)
```

```python
df.head()
```

```python
X3 = df[['species_setosa', 'species_versicolor', 'species_virginica']]
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X3)

kmeans.predict(X3)
```

```python
df['cluster3']=kmeans.predict(X3)
```

```python
sns.scatterplot(data = df, x='petal_length', y= 'petal_width', hue = 'cluster3', palette = 'tab10')
```

```python
sns.scatterplot(data = df, x='sepal_length', y= 'sepal_width', hue = 'cluster3', palette = 'tab10')
```

## Another 3 features

```python
df.head()
```

```python
df['petal_area'] = df.petal_length * df.petal_width
```

```python
df['sepal_area'] = df.sepal_length * df.sepal_width
```

```python
df['sepal_petal_length_ratio'] = df.sepal_length / df.petal_length
```

```python
df['sepal_petal_width_ratio'] = df.sepal_width / df.petal_width
```

```python
X4 = df[['petal_area', 'sepal_area', 'sepal_petal_length_ratio']]
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X4)

kmeans.predict(X4)
```

```python
df['cluster4']=kmeans.predict(X4)
```

```python
sns.scatterplot(data = df, x='petal_length', y= 'petal_width', hue = 'cluster4', palette = 'tab10')
```
